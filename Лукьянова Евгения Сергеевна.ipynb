#### 3. Уберите из названий столбцов знаки переносов ('\n')

```python
# ВАШ КОД ЗДЕСЬ
temp = []
for header in df.columns:
    temp.append(header.replace('\n', ''))
df.columns = temp

df.columns
```

#### 4. Выведите таблицу, которая позволит просмотреть первую дату поставки и общее количество поставленной вакцины в рамках Марки вакцины и страны.

```python
# ВАШ КОД ЗДЕСЬ

# Удаление пустых строк с маркой вакциины
df2 = df.set_index('Марка вакцины')
df3 = df2.drop([np.nan],axis = 0)
df3 = df3.reset_index()

# Сортировка по стране и дате
df4 = df3.sort_values(by = ['Страна','Дата поставки/новости'])

# Выбор первой даты поставки по вакцине
df5 = df4.drop_duplicates(subset = ['Марка вакцины', 'Страна'], keep = 'first', inplace = False)

# Вернемся к исходной сортировке и выберем нужные столбцы
df5 = df5.sort_index(ascending=True)
res = df5[['Марка вакцины', 'Страна', 'Дата поставки/новости', 'Вакцина (кол-во)']]
```

#### 6. Достать домены источников 1, 2, 3
Для столбцов Источников создать столбцы Доменов, где будет извлечен только домен (источник: https://sputnik-abkhazia.ru/Abkhazia/20200320/..., домен: sputnik-abkhazia.ru). В исходном датасете 3 столбца источника и соответственно должно быть создана 3 столбца с доменами.

```python
# ВАШ КОД ЗДЕСЬ
t_source = df[['Источник 1', 'Источник 2', 'Источник 3']]
for source in t_source.columns:
    for id in range(len(t_source.index)):
        if type(t_source[source][id]) != float:
            k = t_source[source][id].split("//")[-1].split("/")[0]
            t_source.at[id, source] = k

df['Домен 1'] = t_source['Источник 1']
df['Домен 2'] = t_source['Источник 2']
df['Домен 3'] = t_source['Источник 3']

df
```
